## Catalysts generation

### R1 list

```python
r1 = ['Cc1ccccc1', 'Cc2c1ccccc1cc3ccccc23', 'Cc1cc(C(C)(C)C)cc(C(C)(C)C)c1', 'Cc1ccc(C(F)(F)F)cc1', 'Cc1ccccn1', 'Cc1ccccc1(C#N)']
r1_names = ['ph', 'ar', 'tbu', 'cf3', 'py', 'cn']
```

### R2 list
This list include the C atom from the scafold because they have different chirality depending on the group.

```python
r2 = ['[C@H](O)', '[C@@H](NC(=S)Nc1cc(C(F)(F)F)cc(C(F)(F)F)c1)', '[C@@H](Nc1c(Nc2cc(C(F)(F)F)cc(C(F)(F)F)c2)c(=O)c1=O)']
r2_names = ['oh', 'thi', 'sqa']
```

```python
cat_list = []
names_list = []
for i in r1:
    for j in r2:
        cat_list.append('C=C[C@H]4C[N@+]5('+i+')CC[C@H]4C[C@H]5'+j+'c6ccnc7ccc(OC)cc67')
        #print('C=CC4C[N+]5('+i+')CCC4CC5C('+j+')c6ccnc7ccc(OC)cc67')
        
for i in r1_names:
    for j in r2_names:
        names_list.append('cat_' + i + '_' + j)
        
#  These smiles codes might not be optimal, but they will be converted again below/
cat_list
print(names_list)
```

```python
mol_list = []
n_mol = 0
for smiles in cat_list:
    mol = chem.MolFromSmiles(smiles)
    mol = chem.MolFromSmiles(chem.MolToSmiles(mol))
    mol_list.append(mol)
    img_mol = Draw.MolToImage(mol)
    img_mol.save(names_list[n_mol] + '.png')
    print(chem.MolToSmiles(mol))
    n_mol += 1
    
img = Draw.MolsToGridImage(mol_list, molsPerRow=3)
img
```

### Generate all the xyz files

```python
opt_cats = []
n_mol = 0
for mol in mol_list:
    cat_H = chem.AddHs(mol)
    AllChem.EmbedMolecule(cat_H)
    AllChem.MMFFOptimizeMolecule(cat_H)
    opt_cats.append(cat_H)
    chem.MolToXYZFile(cat_H, names_list[n_mol]+'.xyz')
    with chem.SDWriter(names_list[n_mol]+'.sdf') as w:
        w.write(cat_H)
    
    n_mol += 1

img = Draw.MolsToGridImage(opt_cats, molsPerRow=3)
img
```

### Generate pictures with the chiral information of all of them

```python
n_mol=0
for mol in mol_list:
    d = rdMolDraw2D.MolDraw2DCairo(500, 500)
    d.drawOptions().addStereoAnnotation = True
    d.drawOptions().addAtomIndices = False
    d.DrawMolecule(mol)
    d.FinishDrawing()
    d.WriteDrawingText(names_list[n_mol]+'.png') 
    n_mol += 1
```
