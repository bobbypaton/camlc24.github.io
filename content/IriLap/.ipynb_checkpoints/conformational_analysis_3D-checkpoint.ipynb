## 2- Conformer generation 

In this first section you will have to select different cutoffs for the conformational analysis.
- **RMSPRE:** pre optimization rmsd threshold. This value will eliminate structures that are too similar to each other before optimizing. You select the threshold. Lower values eliminate only very similar structures while bigger values eliminate more different structures.
- **RMSPOST:** post optimization rmsd threshold. Same but after optimization
- **THREADS:** Number of threads used by the program to perform the calculations
- **ENT:** Energy threshold between from the most stable structure in kcals/mol.
- **PRINTPROPERTY:** If true, includes different properties (Conforer ID, conformer energy and number of rotational bonds) in the output file
- **PRINTXYZ:** If true, prints all the conformers XYZ in individual files. The Conformers are saved with the name conf_*ID*.xyz.

```python
rmspre = 0.1   
rmspost = 0.1
ent = 10        
threads = 3    
printproperty = True
printxyz = True
```

```python
writer = Chem.SDWriter(molecule_out)
suppl = Chem.SDMolSupplier(molecule_sdf) 
```

### Function definitions that will be used later on

```python
def genConf(m, nc, rms, efilter, rmspost):
    nr = int(AllChem.CalcNumRotatableBonds(m))
    m = Chem.AddHs(m)
    Chem.AssignAtomChiralTagsFromStructure(m, replaceExistingTags=True)
    if nc == "X":
        if nr < 3:
            nc = 50
        else:
            nc = nr**3
    ids=AllChem.EmbedMultipleConfs(m, numConfs=nc, pruneRmsThresh=rms, randomSeed=1, useExpTorsionAnglePrefs=True, useBasicKnowledge=True)
    diz = []
    diz2 = []
    diz3 = []
    for id in ids:		
        prop = AllChem.MMFFGetMoleculeProperties(m, mmffVariant="MMFF94s")
        ff = AllChem.MMFFGetMoleculeForceField(m, prop, confId=id)
        ff.Minimize()
        en = float(ff.CalcEnergy())
        econf = (en, id)
        diz.append(econf)	
    if efilter != "Y":
        n, diz2 = ecleaning(m, diz, efilter)
    else:
        n = m
        diz2 = diz
    if rmspost != False and n.GetNumConformers() > 1:
        o, diz3 = postrmsd(n, diz2, rmspost) 
    else:
        o = n
        diz3 = diz2
    return o, diz3, nr
```

```python
def ecleaning(m, diz, efilter):
    diz.sort()
    mini = float(diz[0][0])
    sup = mini + efilter
    n = Chem.Mol(m)
    n.RemoveAllConformers()
    n.AddConformer(m.GetConformer(int(diz[0][1])))
    nid = []
    ener = []
    nid.append(int(diz[0][1]))
    ener.append(float(diz[0][0]))	
    del diz[0]	
    for x,y in diz:
        if x <= sup:
            n.AddConformer(m.GetConformer(int(y)))
            nid.append(int(y))
            ener.append(float(x))
        else:
            break
    diz2 = zip(ener, nid)
    return n, diz2
```

```python
def postrmsd(n, diz2, rmspost):
    diz2 = list(diz2)
    diz2.sort()
    o = Chem.Mol(n)
    o.RemoveAllConformers()
    #confidlist = [i[1] for i in diz2[:]]
    #enval = [i[0] for i in diz2[:]]
    confidlist = []
    enval = []
    nh = Chem.RemoveHs(n)
    del diz2[0]
    for z,w in diz2:
        confid = int(w)
        p=0
        for conf2id in confidlist:
            rmsd = AllChem.GetBestRMS(nh, nh, prbId=confid, refId=conf2id)
            if rmsd <= rmspost:
                p=p+1
                break
        if p == 0:
            confidlist.append(int(confid))
            enval.append(float(z))
    for id in confidlist:
        o.AddConformer(n.GetConformer(id))
    diz3  = zip(enval, confidlist)
    return o, diz3
```

### Conformational analysis calculations

This section performs the conformational analysis calling the previous functions

```python
f = open(molecule_csv, "w")
f.write('ConfId,ConfEnergy,NRotBond\n')
with futures.ProcessPoolExecutor(max_workers=threads) as executor:
    jobs = []
    nm = []
    numMol=0
    for mol in suppl:
        numMol = numMol+1
        if mol is not None:
            nm.append(mol.GetProp('_Name'))
            nc = "X"	

            job = executor.submit(genConf, mol, nc, rmspre, ent, rmspost)
            jobs.append(job)
        else:
            print("ERROR: Impossible to generate conformers for molecule number " + str(numMol))
    widgets = ["Generating conformations; ", progressbar.Percentage(), " ", progressbar.ETA(), " ", progressbar.Bar()]
    pbar = progressbar.ProgressBar(widgets=widgets, maxval=len(jobs))
    for job in pbar(futures.as_completed(jobs)):
        mol,ids,nr = job.result()
        for j in range(0, len(jobs)):
            mol,ids,nr = jobs[j].result()
            name = nm[j]
            for en,id in ids:
                mol.SetProp('_Name', name)
                if printproperty == True:
                    mol.SetProp('ConfId', str(id))
                    mol.SetProp('ConfEnergies', str(en)+ " Kcal/mol")
                    mol.SetProp('Rotatable Bonds Number', str(nr))
                    f.write(str(id) + ',' + str(en) + ',' + str(nr) + '\n')
                
                if printxyz == True:
                    Chem.MolToXYZFile(mol, folder + '/' + 'cat_'+str(id)+'.xyz', confId=id)
                    
                writer.write(mol, confId=id)
writer.close()
f.close()
```

## Population analysis

In this section you will analyze the data and plot the results

### Thermodynamic conditions set up
Set the temperature and the R constant for the energy unit used (*kcal/mol by default*)

```python
R = 0.0019882
T = 298
RT = R * T
```

This sections extracts and analyzes different energy values in order to obtain the population and the relative population of each generated conformer.

- **Index** is the ID number for an specific conformer.
- **Rel_index** is the relative index of an specific conformer, shown as a %.
- **dE** is the relative energy of the conformer, relative to the most stable one.
- **$\exp^{\frac{-ddE}{RT}}$** is the Bolztman population formula.
- **Population** is the population of that conformer shown as %.
- **Cummulative population** is the acummulated population since the most stable conformation until this conformer.

```python
rdkit_db = pd.read_csv(molecule_csv)
rdkit_db['index'] = list(range(1,len(rdkit_db.index)+1))
rdkit_db['rel_index'] = rdkit_db['index'] / len(rdkit_db) * 100
rdkit_db['dE'] = rdkit_db['ConfEnergy'] - rdkit_db['ConfEnergy'][0]
rdkit_db['exp(-ddE/RT)'] = np.exp(-rdkit_db['dE'] / RT)
rdkit_db['Population'] = rdkit_db['exp(-ddE/RT)'] / rdkit_db['exp(-ddE/RT)'].sum() * 100
rdkit_db['Cumulative population'] = rdkit_db['Population'].cumsum()
rdkit_db.to_csv(molecule_csv, index=False)

goal1 = False
for i in range(0,len(rdkit_db)):
    a = rdkit_db['Cumulative population'][i]
    if a >= 90 and not goal1:
        per90 = i + 1
        goal1 = True
    if goal1:
        break
rdkit_db
```

This section creates a plot showing the energy of all the different conformers (<span style="color:blue">**blue**</span>) and the acummulated population at each conformer (<span style="color:red">**red**</span>). It also shows a big dot at the energy value when the population has already reached 90% (<span style="color:darkgreen">**green**</span>).

It saves two different pictures. One in PNG format with transparent background and one in EPS format, which is better if you want to scale the picture.

```python
fig, ax1 = plt.subplots(figsize=(8,5))

matplotlib.rcParams.update({'font.size': 15})

color = '#cc0000'
ax1.set_xlabel('%Conformer')
ax1.set_ylabel('Relative Energy (kcal/mol)', color=color)
ax1.plot(rdkit_db['rel_index'], rdkit_db['dE'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.scatter(rdkit_db['rel_index'][per90-1], rdkit_db['dE'][per90-1], marker='o', color='darkgreen', s=50)

ax2 = ax1.twinx()

color = '#006bff'
ax2.set_ylabel('Cumulative Population (%)', color=color)
ax2.plot(rdkit_db['rel_index'], rdkit_db['Cumulative population'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.ylim(0,100)
fig.tight_layout()
plt.show()

name = molecule_csv.split('.csv')
fig_name = name[0] + '_population_dot.eps'
fig.savefig(fig_name, transparent=True, format='eps')
fig_png = name[0] + '_population_dot.png'
fig.savefig(fig_png, transparent=True, format='png')
```
